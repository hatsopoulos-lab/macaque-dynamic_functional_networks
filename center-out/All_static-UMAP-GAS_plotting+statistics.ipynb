```python
all_trials = [loadPickle('processed_data/rs1050225_trials_by_cutoff.pkl'),
              loadPickle('processed_data/rj031209_trials_by_cutoff.pkl')]
```

```python
afont = {'fontname':'Roboto','fontsize' : 18}
hfont = {'fontname':'Roboto','fontsize' : 20, 'weight': 'bold'}
x_label_coords = [0.12,0.08]
y_label_coords = [-0.03,0.28]

sns.set(font_scale=1.5,style='white',font='Roboto')
num_subjects = 2
subject_names = ['Monkey Rs','Monkey Rj']
fig = plt.figure(figsize=(20,10),facecolor='white')

parent_ax = fig.add_gridspec(ncols=2, nrows=1, width_ratios=[1,1])

cutoff = 1.5
panel_num = 0
panel_num_lower = 4
###############
for sub in range(num_subjects):
    cols = 2
    rows = 2 # num measures

    widths = [1,1]
    heights = [1,1]

    axs = parent_ax[sub].subgridspec(ncols=cols, nrows=rows, width_ratios=widths,
                              height_ratios=heights,wspace=0.1,hspace=0.2 )

    row = 0
    cmap = sns.color_palette("husl",8)

    ##############################################################################################
    # first column: kinematics
    data = all_data[sub]
    trials = all_trials[sub]
    col=0
    ax = fig.add_subplot(axs[row,col])
    for direction in range(8):
        for trial in range(len(data[direction]['trialData'])):
            if trial in trials[cutoff]['trials'][direction]:
                x = data[direction]['trialData'][trial]['trialXPos']
                y = data[direction]['trialData'][trial]['trialYPos']
                y[np.where(y<150)[0]] = y[np.where(y<150)[0]+1]
                min_len = np.min([len(x),len(y)])
                ax.plot(x[:min_len],y[:min_len],c=cmap[direction],linewidth=1, alpha=0.5)

    ax.set_xlabel('KinArm X (au)',**afont)
    ax.set_ylabel('KinArm Y (au)',**afont)
    ax.set_title(f'{subject_names[sub]}',loc='left',**hfont)
    #adjust y-axis label position
    ax.yaxis.set_label_coords(y_label_coords[0]-0.02,y_label_coords[1]+0.01)
    #adjust x-axis label position 
    ax.xaxis.set_label_coords(x_label_coords[0],x_label_coords[1])
    
    ax.set_yticks([])
    ax.set_xticks([])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.text(-0.1, 1.1, string.ascii_uppercase[panel_num], transform=ax.transAxes, 
        size=20, weight='bold')  
    panel_num+=1
    ##############################################################################################
    # second column: UMAP embedding
    embeddings_2d = all_embeddings[sub]
    FN_flattened = all_FN_flattened[sub]
    col = 1
    ax = fig.add_subplot(axs[row,col])

    for s in range(0,8):
        idx = [i for i,x in enumerate(FN_flattened['FN_info']) if x[0]==s]

        sns.kdeplot(x=embeddings_2d[idx, 0], y=embeddings_2d[idx, 1],shade= True,levels=2, color=cmap[s],thresh=0.1,alpha=0.15,ax=ax)

        colors = [cmap[x[0]] for x in np.array(FN_flattened['FN_info'])[idx]]
        ax.scatter(embeddings_2d[idx, 0], embeddings_2d[idx, 1],c=colors,s= 50, alpha=0.8)

    ax.set_xlabel('UMAP X (au)',**afont)
    ax.set_ylabel('UMAP Y (au)',**afont)

    #adjust y-axis label position
    ax.yaxis.set_label_coords(y_label_coords[0],y_label_coords[1])
    #adjust x-axis label position 
    ax.xaxis.set_label_coords(x_label_coords[0],x_label_coords[1])

    ax.set_yticks([])
    ax.set_xticks([])
    ax.invert_xaxis()
    ax.invert_yaxis()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.text(-0.1, 1.1, string.ascii_uppercase[panel_num], transform=ax.transAxes, 
        size=20, weight='bold')  
    panel_num+=1
    ##############################################################################################
    # third column: GAS
    gas_df = all_gas_df[sub]
    row = 1
    ax = fig.add_subplot(axs[row,:])

    # sns.violinplot(data=gas_df,x='dist', y='align_score', hue="type",split=True,)

    order = ['same','neighboring','opposite']
    hue_order = ['data','null']
    pairs=[(("same","data"),("same","null")),
            (("neighboring","data"),("neighboring","null")),
            (("opposite","data"),("opposite","null")),
            (("same","data"),("neighboring","data")),
            (("same","data"),("opposite","data"))]
    sns.violinplot(data=gas_df,x='dist', y='align_score', hue="type", hue_order=hue_order, palette={"data": "#4f4f4f", "null": "#a1a1a1"},inner="quartile")

    annotator = Annotator(ax, pairs, data=gas_df, x='dist', y='align_score', hue='type', order=order,hue_order=hue_order)
    annotator.configure(test='Mann-Whitney', text_format='star', loc='outside',verbose=2,line_height=0.05)
    annotator.apply_test().annotate(line_offset_to_group = 0,line_offset=0.001)
    
    ax.text(-0.1, 1.1, string.ascii_uppercase[panel_num_lower], transform=ax.transAxes, 
    size=20, weight='bold')  
    panel_num_lower+=1
    ax.set_xticklabels(ax.get_xticklabels(),**afont)
    
    ticks = np.arange(0.25,0.525,0.125)
    labels = [round(x,2) if x==0.25 or x == 0.5 else ''  for x in ticks]
    ax.set_yticks(ticks)
    ax.set_yticklabels(labels,**afont)
        
#     ax.set_ylim([0.25,0.45])
    
    plt.ylabel('alignment score',**hfont)
    plt.xlabel('target distance',**hfont)

    leg = ax.legend(frameon=False, handletextpad=0, handlelength=0,loc='lower left')
    colors = ["#4f4f4f", "#a1a1a1"]
    for h, t in enumerate(leg.get_texts()):
        t.set_color(colors[h])
        t.set_weight("bold")
        t.set_size(18) 
        t.set_ha('right')
    sns.despine(ax=ax)
    
plt.tight_layout()
plt.savefig(pjoin(fig_dir,"static_networks.svg"))
```
